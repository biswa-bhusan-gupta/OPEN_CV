```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
# HARRIS CORNER DETECTION:
Flat_Chess=cv2.imread('Flat_ChessBoard.png')
Flat_Chess=cv2.cvtColor(Flat_Chess,cv2.COLOR_BGR2RGB)
plt.imshow(Flat_Chess)
```

```python
Gray_Flat_Chess=cv2.cvtColor(Flat_Chess,cv2.COLOR_RGB2GRAY)
plt.imshow(Gray_Flat_Chess,cmap='gray')
```

```python
Real_Chess=cv2.imread('Real_ChessBoard.jpg')
Real_Chess=cv2.cvtColor(Real_Chess,cv2.COLOR_BGR2RGB)
plt.imshow(Real_Chess)
```

```python
Gray_Real_Chess=cv2.cvtColor(Real_Chess,cv2.COLOR_RGB2GRAY)
plt.imshow(Gray_Real_Chess,cmap='gray')
```

```python
Gray=np.float32(Gray_Flat_Chess)
Gray
```

```python
# BLOCKSIZE:NEIGHBORHOOD SIZE
Dest=cv2.cornerHarris(src=Gray,blockSize=2,ksize=3,k=0.04)
plt.imshow(Dest)
```

```python
Dest=cv2.dilate(Dest,None)
plt.imshow(Dest)
```

```python
Flat_Chess[Dest>0.01*Dest.max()]=[255,0,0] #RGB
plt.imshow(Flat_Chess)
```

```python
Gray=np.float32(Gray_Real_Chess)
Dest=cv2.cornerHarris(src=Gray,blockSize=2,ksize=3,k=0.04)
Dest=cv2.dilate(Dest,None)
Real_Chess[Dest>0.01*Dest.max()]=[255,0,0] #RGB
plt.imshow(Real_Chess)
```

```python
# SHI_TOMASHI CORNER DETECTION:
Flat_Chess=cv2.imread('Flat_ChessBoard.png')
Flat_Chess=cv2.cvtColor(Flat_Chess,cv2.COLOR_BGR2RGB)
Gray_Flat_Chess=cv2.cvtColor(Flat_Chess,cv2.COLOR_RGB2GRAY)

Real_Chess=cv2.imread('Real_ChessBoard.jpg')
Real_Chess=cv2.cvtColor(Real_Chess,cv2.COLOR_BGR2RGB)
Gray_Real_Chess=cv2.cvtColor(Real_Chess,cv2.COLOR_RGB2GRAY)
```

```python
Corners=cv2.goodFeaturesToTrack(Gray_Flat_Chess,64,0.01,10)
Corners
```

```python
Corners=np.int0(Corners)
Corners
```

```python
for i in Corners:
    x,y=i.ravel()
    cv2.circle(Flat_Chess,(x,y),color=(255,0,0),radius=3,thickness=-1)
plt.imshow(Flat_Chess)
```

```python
# 0.01,10:DEFAULT VALUE
Corners=cv2.goodFeaturesToTrack(Gray_Real_Chess,64,0.01,10)
Corners=np.int0(Corners)

for i in Corners:
    x,y=i.ravel()
    cv2.circle(Real_Chess,(x,y),color=(255,0,0),radius=3,thickness=-1)
plt.imshow(Real_Chess)
```

```python
# CANNY EDGE DETECTION:
# 1.Apply Gaussian Filter to Smooth inorder to remove Noise
# 2.Apply Non-Maximum Supression to get rid of spurious response to Edge Detection 
# 3.Apply Double Thresold to determine Potential Edges
# 4.Track Edge by Hysteresis:Finalize the detection of Edges by supressing all the Edges that are weak and not connected to Stronger Edges
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Image=cv2.imread('Sammy_Face.jpg')
plt.imshow(Image)
```

```python
Edges=cv2.Canny(Image,threshold1=127,threshold2=127)
plt.imshow(Edges)
```

```python
Edges=cv2.Canny(Image,threshold1=0,threshold2=255)
plt.imshow(Edges)
```

```python
Med_Value=np.median(Image)
Med_Value
```

```python
# LOWER THRESOLD TO EITHER 0 OR 70% OF MEDIAN VALUE,WHICHEVER IS GREATER:
Lower=int(max(0,0.7*Med_Value))
# UPPER THRESOLD TO EITHER 130 OR 70% OF MEDIAN VALUE,WHICHEVER IS SMALLER:
Upper=int(min(255,1.3*Med_Value))
```

```python
Edges=cv2.Canny(Image,threshold1=Lower,threshold2=Upper+100)
plt.imshow(Edges)
```

```python
New_Image=cv2.resize(Image1,(1000,400))
plt.imshow(New_Image)
```

```python
h_ratio=0.5
w_ratio=0.5
New_Image=cv2.resize(Image1,(0,0),Image1,w_ratio,h_ratio)
plt.imshow(New_Image)
```

```python
h_ratio=0.5
w_ratio=0.5
New_Image=cv2.resize(Image,(0,0),Image,w_ratio,h_ratio)
plt.imshow(New_Image)
```

```python
New_Image=cv2.flip(Image1,0)
plt.imshow(New_Image)
```

```python
New_Image=cv2.flip(Image1,-1)
plt.imshow(New_Image)
```

```python
New_Image=cv2.flip(Image1,1)
plt.imshow(New_Image)
```
